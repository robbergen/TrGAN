## Load the trained generator and generate images

```python
def make_video(fsgen, vgen, n, z_slow=None, z_fast=None):
    xp = fsgen.xp

    if z_slow is None:
        z_slow = xp.random.uniform(-1, 1, (n, fsgen.z_slow_dim)).astype('f')
    z_slow = Variable(z_slow)

    with chainer.using_config('train', False):
        if z_fast is None:
            z_fast = fsgen(z_slow)
        y = vgen_forward(vgen, z_slow, z_fast)
        y = y.reshape(n, -1, y.shape[1], y.shape[2], y.shape[3])
        y = y.transpose(0, 2, 1, 3, 4)

        z_slow = chainer.cuda.to_cpu(z_slow.data)
        z_fast = chainer.cuda.to_cpu(z_fast.data)

    return y, z_slow, z_fast

def make_conditional_video(fsgen, vgen, n, cond, z_slow=None, z_fast=None):
    xp = fsgen.xp

    if z_slow is None:
        z_slow = xp.random.uniform(-1, 1, (n, fsgen.z_slow_dim)).astype('f')
    z_slow = Variable(z_slow)

    with chainer.using_config('train', False):
        if z_fast is None:
            z_fast = fsgen(z_slow)
        y = vgen_forward_cond(vgen, z_slow, z_fast,cond)
        y = y.reshape(n, -1, y.shape[1], y.shape[2], y.shape[3])
        y = y.transpose(0, 2, 1, 3, 4)

        z_slow = chainer.cuda.to_cpu(z_slow.data)
        z_fast = chainer.cuda.to_cpu(z_fast.data)

    return y, z_slow, z_fast

def vgen_forward(vgen, z_slow, z_fast):
    B, n_z_fast, n_frames = z_fast.shape
    z_fast = F.reshape(F.transpose(
        z_fast, [0, 2, 1]), (B * n_frames, n_z_fast))

    B, n_z_slow = z_slow.shape
    z_slow = F.reshape(F.broadcast_to(F.reshape(
        z_slow, (B, 1, n_z_slow)), (B, n_frames, n_z_slow)),
        (B * n_frames, n_z_slow))

    with chainer.using_config('train', False):
        img_fake = vgen(z_slow, z_fast)
    return chainer.cuda.to_cpu(img_fake.data)

def vgen_forward_cond(vgen, z_slow, z_fast,cond):
    B, n_z_fast, n_frames = z_fast.shape
    z_fast = F.reshape(F.transpose(
        z_fast, [0, 2, 1]), (B * n_frames, n_z_fast))

    B, n_z_slow = z_slow.shape
    z_slow = F.reshape(F.broadcast_to(F.reshape(
        z_slow, (B, 1, n_z_slow)), (B, n_frames, n_z_slow)),
        (B * n_frames, n_z_slow))

    with chainer.using_config('train', False):
        img_fake = vgen(z_slow, z_fast,cond)
    return chainer.cuda.to_cpu(img_fake.data)


def load_model(result_dir, config, model_type, snapshot_path=None):
    model_fn = '{}/{}'.format(result_dir, os.path.basename(config['models'][model_type]['fn']))
    model_name = config['models'][model_type]['name']
    kwargs = config['models'][model_type]['args']
    model = imp.load_source(model_name, model_fn)
    model = getattr(model, model_name)(**kwargs)
    if snapshot_path:
        serializers.load_npz(snapshot_path, model)
    return model


def get_models(result_dir, n_iter):
    config = yaml.load(open(glob.glob('{}/*.yml'.format(result_dir))[0]))
    fsgen = load_model(result_dir, config, 'frame_seed_generator', '{}/gen_iter_{}.npz'.format(result_dir, n_iter))
    vgen = load_model(result_dir, config, 'video_generator', '{}/vgen_iter_{}.npz'.format(result_dir, n_iter))
    vdis = load_model(result_dir, config, 'video_discriminator', '{}/vdis_iter_{}.npz'.format(result_dir, n_iter))
    return fsgen, vgen, vdis


def save_video(y, seed, out_dir='infer', prefix=''):
    y = y.transpose(0, 2, 3, 4, 1)
    n, f, h, w, c = y.shape
    y = y.transpose(1, 0, 2, 3, 4)
    hn = int(np.sqrt(n))
    y = y.reshape(f, hn, hn, h, w, c)
    y = y.transpose(0, 1, 3, 2, 4, 5)
    y = y.reshape(f, hn * h, hn * w, c)
    for i, p in enumerate(y):
        fn = '{}/{}_seed-{}_{}.png'.format(out_dir, prefix, seed, i)
        cv.imwrite(fn, p[:, :, ::-1])
    fn = '{}/{}_seed-{}.avi'.format(out_dir, prefix, seed)
    subprocess.call([
        'ffmpeg', '-i', '{}/{}_seed-{}_%d.png'.format(out_dir, prefix, seed),
        '-vcodec', 'rawvideo', '-pix_fmt', 'yuv420p', fn])
    # subprocess.call([
    #     'ffmpeg', '-i', '{}.avi'.format(os.path.splitext(fn)[0]),
    #     '-vcodec', 'libx264', fn.replace('.avi', '.mp4')])
    for _fn in glob.glob('{}/{}_*.png'.format(out_dir, prefix)):
        os.remove(_fn)
    # os.remove('{}.avi'.format(os.path.splitext(fn)[0]))
```

## Define dice scores for segmentations

```python
def calc_dice(a,b):
    #a and b are N-D binary images
    c = a.astype('int')+b.astype('int')
    tf = np.sum(c[c==1])
    tt = np.sum(c[c==2])
    
    return tt/(tf+tt)

def dice_bias(a,b):
    #a and b are N-D binary images
    c = a.astype('int')-b.astype('int')
    tf = np.sum(c[c==1])
    ft = np.sum(c[c==-1])
    return (tf,-ft)
```

```python
#mean over image excluding mask
def mean3(img,val=10**-10):
    #img=img.astype('float')
    masked = np.mean(np.ma.masked_where(img < val, img))
    return masked

#Lesion volume (cm^3)
def MTV(ct_gtvt,pixdim):
    res = np.prod(pixdim)
    return np.sum(np.sum(np.sum(ct_gtvt)))*res/1000

#SUV mean calculation
def SUVmean(pt,ct_gtvt):
    #pt_transformed = registerPET(ct_gtvt,pt)
    #pt_transformed = zoom(pt_transformed,[4,4,1])
    pt_masked = pt*ct_gtvt
    return mean3(pt_masked)

def SUVmax(pt,ct_gtvt):
    pt_masked = pt*ct_gtvt
    return np.max(pt_masked)

#Generate indices for an ROI with approx 1.2 cm diameter
def genROI(shape,coord,pixsize):
    #shape: image dims
    #coord: 3d coord of ROI center
    #pixsize: pixel size in mm
    #print(shape,coord,pixsize)
    bounds = np.ceil(np.array([12,12,12])/pixsize/2).astype('int')
    ind_i = list(range(np.max([coord[0]-bounds[0],0]),np.min([coord[0]+bounds[0]+1,shape[0]])))
    ind_j = list(range(np.max([coord[1]-bounds[1],0]),np.min([coord[1]+bounds[1]+1,shape[1]])))
    ind_k = list(range(np.max([coord[2]-bounds[2],0]),np.min([coord[2]+bounds[2]+1,shape[2]])))
    return(ind_i,ind_j,ind_k)
    

def SUVpeak(pt_transformed,ct_gtvt,pixsize,filt=[12,12,12]):
    #Default filter size of 1.2 cm ^3
    #pixsize is pixel dimensions in mm of CT mask
    pt_masked = pt_transformed*ct_gtvt
    
    #ROI
    sh = np.shape(pt_transformed)

    #convolultion with ROI shaped filter
    filtersize = np.ceil(np.array(filt)/pixsize/2).astype('int')
    nelem = np.prod(filtersize)
    c = np.empty(sh)
    b = np.ones(shape=filtersize)/nelem
    c= convolve(pt_masked, b, mode='constant')

    ind = np.argmax(c)
    ind = np.unravel_index(ind,np.shape(c))
    
    #return SUV peak and the coordinates of the ROI where it was calculated
    return c[ind],ind

# def GLCM2(pt,ct_gtvt): #need to downgrade python for this
#     glcm = RadiomicsGLCM(pt,ct_gtvt)
#     return glcm.getJointEntropyFeatureValue()


def GLCM(pt,ct_gtvt):
    def homog(g):
        h = 0
        for i in range(np.shape(g)[0]):
            for j in range(np.shape(g)[1]):
                h = h+ g[i,j]/(1+np.abs(i-j))
        return h
            
    pt_rescaled = (pt-np.min(pt))*255/np.max(pt-np.min(pt)) #rescale to 0-255
    
    xmin=xmax=ymin=ymax=zmin=zmax = 0
    #find bounding box of lesion
    dim = np.shape(ct_gtvt)[0] #assumes square image
    for i in range(dim):
        if ((np.max(ct_gtvt[i,:,:])>0) & (xmin==0)):
            xmin=i
        if ((np.max(ct_gtvt[dim-1-i,:,:])>0) & (xmax==0)):
            xmax=dim-1-i
        if ((np.max(ct_gtvt[:,i,:])>0) & (ymin==0)):
            ymin=i
        if ((np.max(ct_gtvt[:,dim-1-i,:])>0) & (ymax==0)):
            ymax=dim-1-i
        if ((np.max(ct_gtvt[:,:,i])>0) & (zmin==0)):
            zmin=i
        if ((np.max(ct_gtvt[:,:,dim-1-i])>0) & (zmax==0)):
            zmax=dim-1-i
    patch = pt_rescaled[xmin:xmax,ymin:ymax,zmin:zmax].astype('int')
    
    total = total2 = total3 = 0 #Doing this slice by slice for now. Need to find a 3D version
    energy = energy2 = energy3 = 0
    homo = homo2 = homo3 = 0
    for i in range(np.shape(patch)[2]):
        glcm = greycomatrix(patch[:,:,i], distances=[1], angles=[0,45,90,135,180,225,270,315], levels=256,symmetric=True, normed=True) #P(i,j,d,theta)
        glcm = np.squeeze(np.sum(glcm,axis=-1))/8
        total = total + np.sum(-glcm*np.log2(glcm+10**-8))/np.shape(patch)[2]
        energy = energy + np.sum(glcm*glcm)/np.shape(patch)[2]
        homo = homo + homog(glcm)/np.shape(patch)[2]
    for i in range(np.shape(patch)[1]):
        glcm2 = greycomatrix(patch[:,i,:], distances=[1], angles=[0,45,90,135,180,225,270,315], levels=256,symmetric=True, normed=True) #P(i,j,d,theta)
        glcm2 = np.squeeze(np.sum(glcm2,axis=-1))/8
        total2 = total2 + np.sum(-glcm2*np.log2(glcm2+10**-8))/np.shape(patch)[1]
        energy2 = energy2 + np.sum(glcm2*glcm2)/np.shape(patch)[1]
        homo2 = homo2 + homog(glcm2)/np.shape(patch)[1]
    for i in range(np.shape(patch)[0]):
        glcm3 = greycomatrix(patch[i,:,:], distances=[1], angles=[0,45,90,135,180,225,270,315], levels=256,symmetric=True, normed=True) #P(i,j,d,theta)
        glcm3 = np.squeeze(np.sum(glcm3,axis=-1))/8
        total3 = total3 + np.sum(-glcm3*np.log2(glcm3+10**-8))/np.shape(patch)[0]
        energy3 = energy3 + np.sum(glcm3*glcm3)/np.shape(patch)[0]
        homo3 = homo3 + homog(glcm3)/np.shape(patch)[0]
    
    entropy = (total+total2+total3)/3
    en = (energy+energy2+energy3)/3
    hom = (homo + homo2+ homo3)/3
    
    return (en,hom,entropy)



def noise_params(pt_img,filtsize=7,thresh = 0.25):
    c1 = convolve(pt_img, np.ones(shape=[filtsize,filtsize,filtsize])/filtsize**3, mode='reflect') #E[X]
    c2 = convolve(pt_img*pt_img, np.ones(shape=[filtsize,filtsize,filtsize])/filtsize**3, mode='reflect') #E[X^2]
    
    std = np.sqrt(c2-c1*c1) #Stds of sliding window

    std[c1<thresh] = np.inf

    
    
    buffer = int(filtsize/2)
    stdbuffer=std[buffer:-buffer-1,buffer:-buffer-1,buffer:-buffer-1]
    
    loc = np.argmin(stdbuffer)
    loc = np.unravel_index(loc,np.array(np.shape(stdbuffer)))+np.array([3,3,3])
    
    stdmin = np.std(pt_img[loc[0]-buffer:loc[0]+buffer,loc[1]-buffer:loc[1]+buffer,loc[2]-buffer:loc[2]+buffer].flatten())
    skewness = skew(pt_img[loc[0]-buffer:loc[0]+buffer,loc[1]-buffer:loc[1]+buffer,loc[2]-buffer:loc[2]+buffer].flatten())
    kurt = kurtosis(pt_img[loc[0]-buffer:loc[0]+buffer,loc[1]-buffer:loc[1]+buffer,loc[2]-buffer:loc[2]+buffer].flatten())
    return stdmin,skewness,kurt
import math
def shape_params(ct_gtvt_img,pixdim):
    pi = math.pi
    V = MTV(ct_gtvt_img,pixdim)#volume
    verts,faces,_,_ = marching_cubes_lewiner(ct_gtvt_img,0.99)#surface
    S = mesh_surface_area(verts,faces)/100 #cm^2
   
    compactness1 = V/np.sqrt(pi)/np.sqrt(S**3);
    compactness2 = 36*pi*V**2/S**3;
    sphericity = (36*pi*V**2)**(1/3)/S;
    sphericalDisp = S / (36*pi*V**2)**(1/3);
    Asphericity = (S**3 / V**2 / 36 / pi)**(1/3) - 1; 
    SVratio = S/V;
    Scircle = (3*V/(4*pi))**(2/3)*4*pi;
    Irregularity = S/Scircle;
    
    keys = ('S','compactness1','compactness2','sphericity','sphericalDisp','Asphericity','SVratio','Scircle','Irregularity')
    vals = (S,compactness1,compactness2,sphericity,sphericalDisp,Asphericity,SVratio,Scircle,Irregularity)
    shapedict = dict(zip(keys, vals)) 
    return shapedict

def write_pt(pt,ct,i):
    nifti_img = nib.Nifti1Image(pt, np.eye(4))
    write_path = 'C:\\Users\\robbe\\Downloads\\hecktor_train_v3\\hecktor_train\\synthetic\\SYN'+"{:03d}".format(i)+"\\SYN{:03d}".format(i)+'_pt.nii.gz'
    
    if not os.path.exists('C:\\Users\\robbe\\Downloads\\hecktor_train_v3\\hecktor_train\\synthetic\\SYN'+"{:03d}".format(i)):
        os.makedirs('C:\\Users\\robbe\\Downloads\\hecktor_train_v3\\hecktor_train\\synthetic\\SYN'+"{:03d}".format(i))
    
    nifti_img2 = nib.Nifti1Image(ct, np.eye(4))
    write_path2 = 'C:\\Users\\robbe\\Downloads\\hecktor_train_v3\\hecktor_train\\synthetic\\SYN'+"{:03d}".format(i)+"\\SYN{:03d}".format(i)+'_ct_gtvt.nii.gz'
    
    nib.save(nifti_img, write_path)
    nib.save(nifti_img2, write_path2)
```

```python
import imageio
def write_gif(image,filename):
    im_list = []
    m = np.max(image)
    for i in range(np.shape(image)[2]):
        if m < 2:
            im_list.append(np.squeeze(image[:,:,i].astype('uint8')))
        else:
            im_list.append(np.squeeze(np.sqrt(image[:,:,i]/m*30000).astype('uint8')))
    kargs = { 'duration': 0.33 }
    imageio.mimwrite(filename,im_list,'GIF',**kargs)

i=4
syn_img = np.transpose(y,[3,4,2,0,1])[:,:,:,i,0] #(x,y,z,N,1)
real_img = reals[:,:,:,i]
real_label = masks[:,:,:,i]*(masks[:,:,:,i]>120)


write_gif(syn_img, 'gifs/cond'+ str(i) +'.gif')
write_gif(real_img,'gifs/real'+ str(i) +'.gif')
write_gif(real_label,'gifs/label'+ str(i) +'.gif')
```

```python
import json
def make_splits(prop_fake,val_size,subset=False,subset_indices=(0,1)):
    current_dir =  'C:/Users/robbe/Downloads/hecktor_train_v3/hecktor_train/hecktor_nii_cropped'
    
    if subset:
        sub_real_patients = [p for p in os.listdir(current_dir)][subset_indices[0]:subset_indices[1]+1]
        sub_fake_patients = [p for p in os.listdir(current_dir)][subset_indices[0]+201:subset_indices[1]+202]
        rest_real_patients = [p for p in os.listdir(current_dir)][:subset_indices[0]]+[p for p in os.listdir(current_dir)][subset_indices[1]+1:200]
        rest_fake_patients = [p for p in os.listdir(current_dir)][201:subset_indices[0]+201]+[p for p in os.listdir(current_dir)][201+subset_indices[1]+1:]
        
        #Randomly choose 15 images for validation within subset
        val_set = np.random.choice(range(len(sub_real_patients)), size=(15,), replace=False)    
        ind = np.zeros(len(sub_real_patients), dtype=bool)
        ind[val_set] = True
        rest = ~ind
        
        #Randomly choose 50 images for validation outside subset
        val_set2 = np.random.choice(range(len(rest_real_patients)), size=(50,), replace=False)    
        ind2 = np.zeros(len(rest_real_patients), dtype=bool)
        ind2[val_set2] = True
        rest2 = ~ind2

        sub_real_patients = pd.array(sub_real_patients)
        sub_fake_patients = pd.array(sub_fake_patients)
        rest_real_patients = pd.array(rest_real_patients)
        rest_fake_patients = pd.array(rest_fake_patients)

        sub_train_list = list(sub_real_patients[rest])
        sub_train_fake_list = list(sub_fake_patients[rest])
        sub_val_list = list(sub_real_patients[ind])
        sub_val_list2 = list(rest_real_patients[ind2])
        fake_train_list = list(rest_fake_patients[rest2])
        
        mixed_train = sub_train_list+fake_train_list
        print("Without synthetic data:")
        print("Within-set validation:", "{\"train\": " ,json.dumps(sub_train_fake_list),", \"val\":", json.dumps(sub_val_list),"}")
        print("Without-set validation:", "{\"train\": " ,json.dumps(sub_train_list),", \"val\":", json.dumps(sub_val_list2),"}")
        print("\nWith synthetic data:")
        print("Within-set validation:", "{\"train\": " ,json.dumps(mixed_train),", \"val\":", json.dumps(sub_val_list),"}")
        print("Without-set validation:", "{\"train\": " ,json.dumps(mixed_train),", \"val\":", json.dumps(sub_val_list2),"}")
    else:
        real_patients = [p for p in os.listdir(current_dir)][:200]
        fake_patients = [p for p in os.listdir(current_dir)][201:]

        #Choose validation set randomly
        val_set = np.random.choice(range(len(real_patients)), size=(val_size,), replace=False)    
        ind = np.zeros(len(real_patients), dtype=bool)
        ind[val_set] = True
        rest = ~ind

        real_patients = pd.array(real_patients)
        fake_patients = pd.array(fake_patients)

        train_list = list(real_patients[rest])
        val_list = list(real_patients[ind])
        fake_train_list = list(fake_patients[rest])
        fake_val_list = list(fake_patients[ind])

        #From remaining training set, choose random fake and real images
        val_set2 = np.random.choice(range(len(train_list)), size=(int(len(train_list)*prop_fake),), replace=False)    
        ind2 = np.zeros(len(train_list), dtype=bool)
        ind2[val_set2] = True
        rest2 = ~ind2

        mixed_train = list(pd.array(train_list)[rest2])+list(pd.array(fake_train_list)[ind2])
        print("Without synthetic data:")
        print("{\"train\": " ,json.dumps(list(pd.array(train_list)[rest2])),", \"val\":", json.dumps(val_list),"}")
        print("\nWith synthetic data:")
        print("{\"train\": " ,json.dumps(mixed_train),", \"val\":", json.dumps(val_list),"}")
    
    
        
```
